* Mostrar cuantos papeles para actrices hubo en la década de los 50's

```python
import pandas as pd

titulos= pd.read_csv('imdb_titulos.csv')
titulos.head()


elenco= pd.read_csv('Sin confirmar 563912.crdownload')
elenco.head()

print('Primeros 5 registros del Titulos:')
print(titulos.head(), '\n')

print('Primeros 5 registros del Elenco:')
print(elenco.head(), '\n')

print('Numero de registros en Titulos:', len(titulos))
print('Numero de registros en Elenco:', len(elenco))

print('5 peliculas mas antiguas')

dracula = titulos[titulos["title"].str.contains("Dracula", case=False, na=False)]
print("Películas con la palabra 'Dracula':")
print(dracula)
print("Total de películas con 'Dracula':", len(dracula), "\n")

print("Los 10 títulos más comunes:")
print(titulos["title"].value_counts().head(10), "\n")

romeo = titulos[titulos["title"] == "Romeo and Juliet"].sort_values("year").head(1)
print("Primera película titulada 'Romeo and Juliet':")
print(romeo, "\n")

exorcist = titulos[titulos["title"].str.contains("Exorcist", case=False, na=False)].sort_values("year")
print("Películas con la palabra 'Exorcist' (ordenadas por año):")
print(exorcist, "\n")

print("Películas hechas en 1950:", len(titulos[titulos["year"] == 1950]), "\n")

print("Películas hechas entre 1950 y 1959:", len(titulos[(titulos["year"] >= 1950) & (titulos["year"] <= 1959)]), "\n")

godfather_roles = elenco[elenco["title"] == "The Godfather"]
print("Roles en 'The Godfather':")
print(godfather_roles)
print("Total de roles:", len(godfather_roles), "\n")

dracula_1958 = elenco[(elenco["title"] == "Dracula") & (elenco["year"] == 1958)].sort_values("n")
print("Elenco de 'Dracula' (1958) ordenado por n:")
print(dracula_1958, "\n")

bruce_wayne = elenco[elenco["character"] == "Bruce Wayne"]
print("Cantidad de papeles de 'Bruce Wayne':", len(bruce_wayne), "\n")

de_niro = elenco[elenco["name"] == "Robert De Niro"]
print("Cantidad de papeles de Robert De Niro:", len(de_niro), "\n")

heston_60s = elenco[(elenco["name"] == "Charlton Heston") & (elenco["n"] == 1) &
                    (elenco["year"].between(1960, 1969))].sort_values("year", ascending=False)
print("Papeles protagonistas de Charlton Heston en los 60's:")
print(heston_60s, "\n")

actores_50s = elenco[(elenco["type"] == "actor") & (elenco["year"].between(1950, 1959))]
print("Papeles para actores en los 50's:", len(actores_50s), "\n")

actrices_50s = elenco[(elenco["type"] == "actress") & (elenco["year"].between(1950, 1959))]
print("Papeles para actrices en los 50's:", len(actrices_50s), "\n")



```
